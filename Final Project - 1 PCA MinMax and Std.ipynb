```python
# Load Data
from scipy.io import arff
import pandas as pd

data_files = [
    "dataset/Scenario/Scenario B/TimeBasedFeatures-Dataset-15s-AllinOne.arff"
]
iscx_tbf = None

for data_file in data_files:
    print(f"{data_file}")
    with open(data_file, "r") as f:
      data, meta = arff.loadarff(f)
      ndf = pd.DataFrame(data=data, columns=meta.names())
      if iscx_tbf is None:
        iscx_tbf = ndf
      else:
        iscx_tbf = pd.concat([iscx_tbf, ndf])
    
# fix a loadarff limitation with strings
iscx_tbf["class1"] = iscx_tbf["class1"].map(lambda x: x.decode("utf-8"))

iscx_tbf.head(5)
```

# Try minmax scaler

```python
from sklearn.preprocessing import MinMaxScaler

# Features obtained from `MinMaxScaler`
iscx_features_minmax = pd.DataFrame()

for col_name in iscx_features.columns:
    # 1) Convert column data to NumPy array and add an extra dimension
    # 2) Re-scale data with `MinMaxScaler`
    # 3) Update column with scaled result
    
    narray = iscx_features[col_name].to_numpy().reshape(-1, 1)
    scaler = MinMaxScaler().fit(narray)
    narray = scaler.transform(narray)
    narray = narray.squeeze()
    iscx_features_minmax[col_name] = narray

# Convert features to NumPy array
iscx_features_minmax = iscx_features_minmax.values
iscx_y_minmax = y.values
```

# Try Std Scaler

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Features obtained from `StandardScaler`
iscx_features_std = pd.DataFrame()

# Remove outliers for `StandardScaler`
remove_outliers = True
# Outlier threshold (standard deviation as unit)
outlier_threshold = 10
# Indices of records with outlier values
outlier_indices = set()

for col_name in iscx_features.columns:
    # Obtain column data as 2D NumPy array
    col_data = iscx_features[col_name].values.astype(float)[:, None]
    # Re-scale data with `StandardScaler`
    col_data = StandardScaler().fit_transform(col_data)
    
    if remove_outliers:
        # 1) Get indices of records with outlier column values
        col_outlier = np.abs(col_data) > outlier_threshold
        col_outlier = col_outlier.squeeze()
        outliers = np.nonzero(col_outlier)
        # 2) Merge these indices with `outlier_indices`
        outlier_indices.update(outliers[0])
        
    # Copy rescaled column
    iscx_features_std[col_name] = col_data.squeeze()

# Remove records with outlier values
if remove_outliers:
    iscx_features_std.drop(outlier_indices, inplace=True)

# Convert features to NumPy array
iscx_features_std = iscx_features_std.values
```

```python
iscx_y_std = y
# Removing labels for samples with outlier data for consistency
if remove_outliers:
    iscx_y_std = y.drop(outlier_indices)

# Convert labels to NumPy array
iscx_y_std = iscx_y_std.values
```

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def visualize_samples(
        samples, labels,
        title="Samples visualization",
        colors=["green", "orange", "red", "blue", "yellow", "magenta", "purple"],
        dimensions=""
    ):
    """ Visualize first three dimensions of samples. """
    if dimensions == "fiat":
        cols = [3, 5, 1]
    elif dimensions == "biat":
        cols = [4, 6, 2]
    elif dimensions == "flowiat":
        cols = [11, 12, 14]
    elif dimensions == "active":
        cols = [15, 17, 18]
    elif dimensions == "other":
        cols = [15, 17, 18]
    else:
        cols = [0, 1, 2]
    # Convert colors to NumPy array
    colors = np.array(colors)
    # Create figure
    fig = plt.figure(figsize=(7, 3))
    
    # Plot first two dimensions on a 2D plot
    ax_2d = fig.add_subplot(1, 2, 2)
    ax_2d.set_title(f"{title} [2D]")
    ax_2d.scatter(samples[:, cols[0]], samples[:, cols[1]], c=colors[labels])
    ax_2d.set_xlabel(iscx_features.columns[cols[0]])
    ax_2d.set_ylabel(iscx_features.columns[cols[1]])
    
    # Plot first three dimensions on a 3D plot
    ax_3d = fig.add_subplot(1, 2, 1, projection="3d")
    ax_3d.set_title(f"{title} [3D]")
    ax_3d.scatter(samples[:, cols[0]], samples[:, cols[1]], samples[:, cols[2]], c=colors[labels])
    ax_3d.set_xlabel(iscx_features.columns[cols[0]])
    ax_3d.set_ylabel(iscx_features.columns[cols[1]])
    ax_3d.set_zlabel(iscx_features.columns[cols[2]])

# Visualize `StandardScaler` features (without dimensionality reduction)
visualize_samples(iscx_features_std, iscx_y_std, title="Standard Scaler")
visualize_samples(iscx_features_minmax, iscx_y_minmax, title = "MinMaxScaler")
```

```python
visualize_samples(iscx_features_std, iscx_y_std, title = "MinMaxScaler", dimensions="fiat")
visualize_samples(iscx_features_std, iscx_y_std, title = "MinMaxScaler", dimensions="biat")
visualize_samples(iscx_features_std, iscx_y_std, title = "MinMaxScaler", dimensions="flowiat")
```

# Kernel PCA

```python
import time
from contextlib import contextmanager

@contextmanager
def timeit(action="Timing"):
    # Record start time
    print(f"{action} started...")
    start_time = time.time()
    
    # Execute task
    yield
    
    # Compute and show elapsed time
    elapsed_time = time.time()-start_time
    print(f"{action} completed. Elapsed time: {elapsed_time:.2f}s")

# Test timing function by sleeping for 1 second
with timeit("Testing timing function"):
    time.sleep(1)
```

```python
from sklearn.decomposition import KernelPCA

def get_KernelPCA(components=3, kernel="poly", data=iscx_features_minmax):
    pca_poly_3 = KernelPCA(n_components=components, kernel=kernel)
    with timeit(f"Kernel PCA (Polynomial; d = {components})"):
        return pca_poly_3.fit_transform(data)
```

```python
# Plot kernel PCA result
minmax_pca_3 = get_KernelPCA(components=3, kernel="poly")
visualize_samples(minmax_pca_3, iscx_y_minmax, title="MinMax Kernel PCA (Polynomial; d = 3)", dimensions="")
```
